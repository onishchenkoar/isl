```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC


sns.set()
%matplotlib inline
```

From the textbook, p. 369:
> Generate a simulated two-class data set with 100 observations and
two features in which there is a visible but non-linear separation between the two classes.

```python
np.random.seed(1)
var1 = np.random.uniform(low=-0.5, high=0.5, size=100)
var2 = np.random.uniform(low=-0.5, high=0.5, size=100)
class_labels = np.where(var2 > 2*var1**2, 'Orange color', 'Blue color')
plt.title('Meaningful plot with a good legend')
ax = sns.scatterplot(x=var1, y=var2, hue=class_labels)
```

> Show that in this setting, a support vector machine with a polynomial kernel (with degree greater than 1) or a radial kernel will outperform a support vector classifier on the training data. Which technique performs best on the test data? Make plots and report training and test error rates in order to back up your assertions.

```python
var1 = var1.reshape(-1, 1)
var2 = var2.reshape(-1, 1)
x = np.concatenate((var1, var2), axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, class_labels)
```

```python
svm_linear = SVC(kernel='linear')
svm_linear.fit(x_train, y_train)
print(f'Linear SVM, train score: {svm_linear.score(x_train, y_train):.2f}')
print(f'Linear SVM, test score: {svm_linear.score(x_test, y_test):.2f}\n')

svm_quadratic = SVC(kernel='rbf')
svm_quadratic.fit(x_train, y_train)
print(f'RBF SVM, train score: {svm_quadratic.score(x_train, y_train):.2f}')
print(f'RBF SVM, test score: {svm_quadratic.score(x_test, y_test):.2f}')
```
